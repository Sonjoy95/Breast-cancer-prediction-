# Breast Cancer Prediction

Breast cancer prediction is a classification problem. we are using the sklearn's inbuilt dataset for the prediction. We are using different classification algorithms like **support vector machine, Dicision tree, Random forest, K-nearest neighbors, Naive bayes classifier & Logistic regression** for the purpose of better performance. 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.datasets as data
```

## (i) Support Vector Classification

```python
dataset=data.load_breast_cancer()
dataset
```

```python
breast_cancer_df=pd.DataFrame(data=dataset.data,columns=dataset.feature_names)
breast_cancer_df["target"]=dataset.target
breast_cancer_df
```

```python
x=breast_cancer_df.drop("target",axis=1)
y=breast_cancer_df["target"]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2020)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.svm import SVC
```

**Applying on unscaled data**

```python
class_rbf=SVC(kernel="rbf")
class_rbf.fit(x_train,y_train)
```

**R2 score**

```python
class_rbf.score(x_test,y_test)
```

**On scaled data**

```python
sc=StandardScaler()
sc.fit(x_train)
x_train_scale=sc.transform(x_train)
x_test_scale=sc.transform(x_test)
```

```python
class_rbf1=SVC(kernel="rbf")
class_rbf1.fit(x_train_scale,y_train)
```

**R2 Score on scaled data**

```python
class_rbf1.score(x_test_scale,y_test)
```

Support vector classifier performed better on scaled data. So we are considering scaled data.
In support vector classifier we can use different **kernels** like **rbf, poly, linear** etc.

```python
class_poly=SVC(kernel="poly",degree=2)
class_poly.fit(x_train_scale,y_train)
```

```python
class_poly.score(x_test_scale,y_test)
```

```python
class_linear=SVC(kernel="linear")
class_linear.fit(x_train_scale,y_train)
```

```python
class_linear.score(x_test_scale,y_test)
```

From above all kernels **linear** is performed better, so we are considering the model in which we have used kernel linear i.e. **class_linear** model.

## (ii) Decision Tree Classification

 Using **SVC** algorithms we have predicted a certain patient has a cancerous tumor or not, we can predict same by using **Decision Tree's Decision Tree Classification** algorithm.
 
 Below we have imported **DecisionTreeClassifier (DTC) function** to apply **Decision Tree Classification** algorithm.

```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
classifier_gini=DecisionTreeClassifier(criterion="gini")
classifier_gini.fit(x_train_scale,y_train)
```

```python
classifier_gini.score(x_test_scale,y_test)
```

Above we have applied **DTC** function with **criterion="gini"** and below we have applied **criterion="entropy"**, we can see that **gini** gives **94%** accuracy and **entropy** gives **92%** accuracy, this means **gini** performed well than **entropy**. So we are considering the **classifier_gini** model for prediction.

```python
classifier_entropy=DecisionTreeClassifier(criterion="entropy")
classifier_entropy.fit(x_train_scale,y_train)
```

```python
classifier_entropy.score(x_test_scale,y_test)
```
